```python
import pandas as pd
import numpy as np
import json
import matplotlib.pyplot as plt
```

```python
path = '../DataCollection/data/'
```

```python
# user 1 data
data = []
for i in range(1,5):
    with open(path + 'user1song{}.json'.format(i)) as json_file:
        data.append(json.load(json_file))
```

```python
df = pd.DataFrame(data)
df
```

```python
fig, ax = plt.subplots(12,1, figsize=(15,40))
for i in range(4):
    ax[3*i].plot(np.arange(0, 120, 1/128), df.iloc[i]['raw_eeg'])
    ax[3*i+1].plot(np.arange(0,120,1/128), df.iloc[i]['attention'])
    ax[3*i+2].plot(np.arange(0,120,1/128), df.iloc[i]['meditation'])
plt.tight_layout()
```
